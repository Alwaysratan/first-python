```python

# Variables


# A variable is used to store data.

name = "Ratan"
age = 21
height = 5.8
is_student = True

print(name)
print(age)
print(height)
print(is_student)
```

```python

# Variable Naming Rules


#  Valid Variable Names

name = "Ratan"
_age = 21
age2 = 22
first_name = "Rahul"
myVariable = 100
PI = 3.14159

print(name)
print(_age)
print(age2)
print(first_name)
print(myVariable)
print(PI)

#  Invalid Variable Names

# 2name = "Ratan"        # Cannot start with a number
# first-name = "Rahul"   # Hyphen (-) is not allowed
# class = "Python"       # Reserved keyword
# my variable = 10       # Space is not allowed

# Rules:
# 1. Must start with a letter or underscore (_)
# 2. Cannot start with a number
# 3. Can contain letters, numbers, and underscores
# 4. Cannot contain spaces or special characters
# 5. Variable names are case-sensitive
```

```python

# Assigning Variables


# Assign different data types

name = "Ratan"
age = 21
salary = 50000.50
is_employee = True

print(name)
print(age)
print(salary)
print(is_employee)
```

```python

# Multiple Assignment


# Assign multiple values

a, b, c = 10, 20, 30

print(a)
print(b)
print(c)

# Same value to multiple variables

x = y = z = 100

print(x)
print(y)
print(z)
```

```python

# Swapping Variables


a = 10
b = 20

print("Before Swapping")
print("a =", a)
print("b =", b)

a, b = b, a

print("After Swapping")
print("a =", a)
print("b =", b)
```

```python

# User Input


name = input("Enter your name: ")

print("Hello", name)
```

```python

# Input with Type Casting


# Integer Input

age = int(input("Enter your age: "))
print(age)
print(type(age))

# Float Input

height = float(input("Enter your height: "))
print(height)
print(type(height))

# String Input

city = input("Enter your city: ")
print(city)
print(type(city))

# Boolean Input

is_student = bool(input("Are you a student? "))
print(is_student)
print(type(is_student))
```
