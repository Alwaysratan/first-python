```python
#datetime
import datetime
```

```python
x=datetime.datetime.now()
print(x)
y=datetime.datetime(1997,10,14)
print(y.strftime("%a"))
print(y.strftime("%A"))
print(y.strftime("%m"))
print(y.strftime("%y"))
print(y.strftime("%S"))
print(y.strftime("%f"))
```

```python
#math
import math
```

```python
x=max(13,65,40) #maximum among them
print("maximum",x)
x=min(13,65,40)#minimum among them
print("minimum",x)
```

```python
a=pow(2,4) #power
print(a)
```
