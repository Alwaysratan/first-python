```python
#functions
# functions are a set of code , which once created ,they can be used throughout the program.

# functions help break our program into smaller parts and helps it look more organized and manageable
```

```python
#defining functions
def hello():
    print("hello world")
hello()

def add():
    x=33
    y=88
    print(x+y)

add()
```

```python
#parameters
#parameters are the variables writtern inside the parenthesis with the name of function.
def add(a,b):
    print(a+b)

def rect(length,width):
    print("area of the rectangle is",length*width)
```

```python
#arguments
#arguments are the values passed to the parameters while calling the function.
add(12,23)
add(74,90)
rect(23,5)
```

```python

# Positional Arguments


def student(name, age):
    print(name, age)

student("Ratan", 22)
```

```python

# Keyword Arguments


def student(name, age):
    print(name, age)

student(age=22, name="Ratan")
```

```python

# Default Arguments


def greet(name="Guest"):
    print("Hello", name)

greet()
greet("Ratan")
```

```python

# Arbitrary Arguments (*args)


# Accepts multiple positional arguments.

def total(*numbers):
    print(sum(numbers))

total(10,20)
total(10,20,30,40)
```

```python


#  Arbitrary Keyword Arguments (**kwargs)


# Accepts multiple keyword arguments.

def student(**details):
    print(details)

student(name="Ratan", age=22, city="Guwahati")
```

```python
#return statement
#return keyword in python is used to exit a funtion and return the value of the function
def hello():
    return("hello world")
print(hello())
```

```python
def add(a,b):
    return(a+b)
print(add(334,44))
```

```python
#recursion
#recursion in most commonly used mathematical and programming concept.
#in simple words,recursion means a function can call itself,giving us a benefit of looping through data in order to get a result.

def hello():
    print("hello")
    
    return(hello())
    
print(hello())
    
```

```python
def fact(n):
    if n==1:
        return 1
    else:
        return(n*fact(n-1))
print(fact(int(input("enter the number"))))
```

```python

# Lambda Functions


square = lambda x: x**2

print(square(6))

add = lambda a,b: a+b

print(add(10,20))
```

```python

# Local & Global Variables
# local variables are restricted to only one block of code and cannot be changed throughout the program.
# global variables are not restricted to one block of code and that can be changed througthout the program.   

x = 100

def demo():
    x = 50
    print("Local:", x)

demo()

print("Global:", x)
```

```python
#problem solving
#write a function to find maximum of three numbers in python.
def max_num(v1,v2,v3):
    if v1>v2 and v1>v3:
        return v1
    elif v2>v1 and v2>v3:
        return v2
    else:
        return v3
print("maximum number is",max_num(1,31,8))
```

```python
#write a python function to create and print a list where the values are square of numbers between 1 to 30.
def create_list():
    l=[]
    for i in range(1,31):
        l.append(i**2)

    return l

print(create_list())
```

```python
#write a function that takes a numbers as a parameter and check if the number is prime ar not.
def check_prime(num):
    if num==1:
        print("it is not a prime number")
    if num==2:
        print("it is a print number")
    if num>2: 
       for i in range (2,num):
           if num %2==0:
               print("it is not a prime number")
               break
       else:
            print("it is a prime number")

check_prime(11)
```

```python
#write a function to sum all the numbers in a list.
def add(num):
    total=0
    for i in num:
        total=total+i
    return(total)
print("sum of all the items in the list is",add([12,43,45,3,4,5,]))
```

```python
#write a program to solve the fibonacci sequence using recursion
def fs(num):
     if num==1:
         return (0)
     elif num==2:
         return(2)
     else:
         return(fs(num-1)+fs(num-2))
print(fs(5))
```
