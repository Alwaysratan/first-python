```python
#sets are unordered collection of data. every element inside the set is unique and mutable.
# sets are written inside the curly brackets{}.
# the value inside a set is sepated by coma,.
# mutable means once created , they can be changed.

## Creating Sets
a= {"ironman","hulk","thor","captain america"}
print(a)
print(type(a))
```

```python
# iteration of sets
for i in a:
    print(i)
```

```python
#set functions part 1
# add()
a.add("spiderman")
print(a)

# remove
a.remove("thor") # remove at specific value from set
print(a)

# discard()
a.discard("ironman")#same like discard
print(a)

# pop()
a.pop(a #random value remove
print(a)

# copy()
b=a.copy()
print(a)
```

```python
# set function 2

a= {"ironman","hulk","thor","captain america"}
b= {"superman","batman","joker"}
c= {"hulk","thor"}
# issubset()
print(c.issubset(a))
# issuperset()
print(a.issuperset(c))
# isdisjoint()
print(a.isdisjoint(b))
# clear()
print(a.clear())
# update()
a.update(c)
print(a)
```

```python
# set function 3
a= {"ironman","hulk","thor","captain america"}
b= {"superman","batman","joker"}
c= {"hulk","thor"}
# union()
print(a.union(c)) # add all the element but does not repeat
# intersection()
print(a.intersection(c)) # give only common one
# difference()
print(a.difference(c)) #can not change in origanal set rather than it can store in another variable
# symmetric_difference()
print(a.symmetric_difference(b)) #ignore the common elements

# changes in origanal set
# intersection_update()
print(a.intersection_update(c))
# difference_update()
print(a.difference_update(c)) 
# symmetric_difference_update()
print(a.symmetric_difference_update(b))
```
